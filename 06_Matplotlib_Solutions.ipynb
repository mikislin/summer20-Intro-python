**(a)** produce a plot of the functions f(x) = e^(-x/10) * arctan(πx) and g(x) = e^(x/10) * cos(πx) over the interval [0, 10]on same axis.

Include labels for the x- and y-axes, and a legend explaining which line is which plot

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,11,0.1)
y = np.exp(-x/10)*np.arctan(np.pi*x)
y1 = np.exp(-x/10)*np.cos(np.pi*x)

fig,ax = plt.subplots()
ax.plot(x,y, color='red', linewidth=2)
ax.plot(x,y1, color='green', linewidth=2)

```

**(b)** The shape of limacon can be defined parametrically as 

r = r0 + cos(a)
x = r * cos(a)
y = r * sin(a)

Use this definition to plot the shape for r0 = 0.2, r0 = 0.5, and r0 = 0.8. Be sure to use enough points that the curve is closed and appears smooth. Use a legend to identify which curve is which.

```
r0s = np.array([0.2, 0.5, 0.8])
a= np.arange(0,360,0.1)
fig, ax = plt.subplots(nrows=len(r0s), ncols=1, sharex=True)

for i in range(len(r0s)):
  r = r0s[i] +np.cos(a)
  x = r*np.cos(a)
  y = r*np.sin(a)
  ax[i].plot(x,y, color='red', linewidth=2)
```

(d) Make sin(x),cos(x), tan(x),-argtan(x) plot with 4 subplots (2 rows, 2 columns)

hint: you'll now have to index into axes like axes[0, 0]).


1.   Create a loop over all the axes objects (hint: use `axes.flatten()`) so that the legend and set_ylabel functions are called for all subplots.
2.   Use the loop from #2 to add a title to only the top row of plots using the `set_title function`.
3.   Look at the documentation for the `fig.tight_layout()` command to optimize figure layout. * Note that this doesn't play nice with the figure suptitle. try using the `plt.subplots_adjust(top=0.85) `command to control the whitespace at the top of the plot.

```
#make a time array
t = np.arange(0, 10, 0.1)

# Create figure and axes objects. Make them share the x-axis
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True)
plt.subplots_adjust(top=0.85)

# Plot on each subplot by indexing into 'ax'
axes[0,0].plot(t, np.sin(t), label='sin')
axes[1,0].plot(t, np.cos(t), label='cos', color='red')

axes[0,1].plot(t, np.tan(t), label='tan')

axes[1,1].plot(t, -np.arctan(t), label='-arctan', color='green')


# We can loop over the `axes` array to set properties in every subplot (no matter how many)
for ax in axes.flatten():
    ax.legend(loc='best')
    ax.set_ylabel('Amplitude', fontsize=14)

axes[0,1].set_ylabel("")
axes[1,1].set_ylabel("")


#put titles on only plots in the 0 row
for col,ax in enumerate(axes[0, :].flatten()):
    ax.set_title('column {}'.format(col))
    
# We can also access individual axes to set the properties
axes[1,0].set_xlabel('Time', fontsize=14, weight='bold', style='italic')
axes[1,1].set_xlabel('Time', fontsize=14, weight='bold', style='italic')
```
